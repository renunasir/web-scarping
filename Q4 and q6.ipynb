# 4)Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htmÂ¶

```python
from bs4 import BeautifulSoup
import requests
```

```python
pre_page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
pre_page
```

```python
pre_page.content
```

```python
soup = BeautifulSoup(pre_page.content)
soup
```

```python
print(soup.prettify())
```

# Name of president

```python
header = soup.find_all(['h3'])
header
```

```python
headings = []
for i in header:
    headings.append(i.text.replace('\n',''))
headings
```

```python
parag = soup.find_all('div',class_='presidentListing')
parag
```

# Name and term

```python
pcom = []
for i in parag:
    pcom.append(i.text.replace('\n',''))
pcom
```

# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape

a)	Top 10 ODI teams in menâs cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
iccm = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
iccm
```

```python
iccm.content
```

```python
soup = BeautifulSoup(iccm.content)
soup
```

```python
ranking_table1 = soup.find_all('tr',class_='rankings-block__banner')
ranking_table1
```

```python
ranking_table = soup.find_all('span',class_='u-hide-phablet')
ranking_table
```

```python
Top_team = []
for i in ranking_table:
    Top_team.append(i.text.replace('\n',''))
    del(Top_team[10:])
Top_team
```

```python
stat_table = soup.find_all('table',class_='table')
stat_table
```

```python
len(stat_table)
```

```python
type(stat_table)
```

```python
stat_table = stat_table[0]
```

```python
type(stat_table)
```

```python
for row in stat_table.find_all('tr'):
    for cell in row.find_all('td'):
        print(cell.text.replace('\n',''))
```

```python
with open ('cricket_stats.txt','w') as r:
    for row in stat_table.find_all('tr'):
        for cell in row.find_all('td'):
            r.write(cell.text.ljust(10))
        r.write('\b')
```
