1.	Scrape the details of most viewed videos on YouTube from Wikipedia.

A)	Rank
B)	Name
C)	Artist
D)	Upload date
E)	Views


```python
# Importing Libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
import re
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from datetime import datetime
```

```python
# first connecting with webdriver
driver=webdriver.Chrome(r"C:\Users\user\Desktop\drivers\chromedriver")
```

```python
# opening website using webdriver 
url = ("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.get(url)
```

```python
#creating empty list
Rank = []
Name = []
Artist = []
Upload_Date = []
views = []
```

```python
#Scraping Rank
for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]'):
            Rank.append(i.text)
Rank=Rank[0:30]

#Scraping name
for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]'):
            Name.append(i.text)
Name=Name[0:30]

#Scraping Artist
for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]'):
            Artist.append(i.text)
Artist=Artist[0:30]

#Scraping Upload Date
Upload_Date=[]
for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]'):
            Upload_Date.append(i.text)
Upload_Date=Upload_Date[0:30]

#Scraping Views
Views=[]
for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]'):
            Views.append(i.text)
Views=Views[0:30]
```

```python
#Creating Dataframe
wiki=pd.DataFrame({})
wiki['Rank']=Rank
wiki['Name[1]']=Name
wiki['Artist']=Artist
wiki['Upload Date']=Upload_Date
wiki['Views']=Views
wiki
```

```python
#Removing stray numbers from names
new = wiki["Name[1]"].str.split("[", n = 1, expand = True) 
new
```

```python
#Dropping the column with stray numbers
wiki.drop(columns=['Name[1]'], axis=1, inplace=True)
```

```python
#Inserting the name column
wiki.insert(1, "Name", new[0])
```

# 6.	Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/

```python
# Importing Libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
import re
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from datetime import datetime
```

```python
# first connecting with webdriver
driver=webdriver.Chrome(r"C:\Users\user\Desktop\drivers\chromedriver")
```

```python
# opening website using webdriver 
url = ("https://www.billboard.com/charts/hot-100")
driver.get(url)
```

```python
#Creating empty lists
Song=[]
Artist=[]
Last_Week_Rank=[]
Peak_Rank=[]
Weeks_on_board=[]

#Scraping Song Name
ntag=driver.find_elements(By.XPATH,"//span[@class='chart-element__information__song text--truncate color--primary']")
for i in ntag:
    if i.text is None :
        Song.append("--") 
    else:
        Song.append(i.text.replace("/"," "))
        

#Scraping Artist
atag=driver.find_elements(By.XPATH,"//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in atag:
    if i.text is None :
        Artist.append("--") 
    else:
        Artist.append(i.text)
        
last_rank_tag=driver.find_elements(By.XPATH,"//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in last_rank_tag:
    if i.text is None :
        Last_Week_Rank.append("--") 
    else:
        Last_Week_Rank.append(i.text)
        
          
#scraping peak rank
peak_tag=driver.find_elements(By.XPATH,"//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in peak_tag:
    if i.text is None :
        Peak_Rank.append("--") 
    else:
        Peak_Rank.append(i.text)
        

#Scraping Weeks on board
wob=driver.find_elements(By.XPATH,"//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in wob:
    if i.text is None :
        Weeks_on_board.append("--") 
    else:
        Weeks_on_board.append(i.text)
```

```python
# creating the dataframe from the scraped data
import pandas as pd
TOP100SONGS=pd.DataFrame({})
TOP100SONGS['Song_name']=Song
TOP100SONGS['Artist_name']=Artist
TOP100SONGS['Last_Week_Rank']=Last_Week_Rank
TOP100SONGS['Peak_Rank']=Peak_Rank
TOP100SONGS['Weeks_on_board']= Weeks_on_board
TOP100SONGS

 
```
